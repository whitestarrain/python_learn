```python
import string

import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame(
    np.arange(25).reshape(5, 5), columns=list(string.ascii_lowercase[:5])
)
df1
```

```python
df2 = pd.DataFrame(np.arange(25).reshape(5, 5), columns=list(
    string.ascii_lowercase[-5:]), index=[1, 2, 3, 4, 5], )
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```
