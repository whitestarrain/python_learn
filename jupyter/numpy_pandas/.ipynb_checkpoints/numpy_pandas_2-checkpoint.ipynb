```python
# join 是根据行index进行合并，行长度变长
# merge是根据列column进行合并，有inner（默认），outer，left，right三种
# 类似数据库操作
df1=pd.DataFrame(np.arange(50).reshape(10,5),columns=list(string.ascii_lowercase[:5]))
print(df1)
print(df2)
df1.merge(df2,how="outer")
# 也可以通过 on 指定根据的column，默认是两个表column的交集
# 相当于 select * from df1 full [outer] join df2 on df1.a=df2.a .....;   (先笛卡尔积，再去重)
```

```python
# 分组，其实通过bool索引就能做到
grouped = sdata.groupby("Country")
# 获得可迭代 分组结果对象
print(grouped)
```

```python
for key,value in grouped:
    print(key,value)
    break
print("-"*100)
print(type(list(grouped)[0]))
print(type(list(grouped)[0][0]))
print(type(list(grouped)[0][1]))
print("-"*100)
list(grouped)[1][1]
```

```python
grouped[["Brand"]]# 这样取到的是DataFrameGroupBy
#  DataFrameGroupBy 取列得到 SeriesGroupBy
grouped["Brand"]
```

```python
# 聚合，用来统计分组个数（NaN会不进行统计）
grouped.count()
```

```python
# 中美各自的数量
print(grouped["Brand"].count()["CN"])
print(grouped["Brand"].count()["US"]) # DataFrameGroupBy-->SeriesGroupBy-->Series-->取值
```

```python
# 求中国各个省份的数量

# 获得中国组对应的数据
group_cn = grouped.get_group("CN")
# 也可以这样做：sdata[sdata["Country"]=="CN"]

group_cn # dataframe类型
```

```python
# 中国每个省的Brand数量（因为Brand一列没有NaN）
group_cn.groupby("State/Province")["Brand"].count() # 转为 SeriesGroupBy 再 聚合
# 两个结果相同
group_cn.groupby("State/Province").count()["Brand"] # 先聚合得到DataFrame再转为 Series
```

```python
print(type(sdata.groupby(by=["Country","State/Province"])["Brand"].count()))
# 上下两行结果相同，都是两行index的series
print(type(sdata["Brand"].groupby(by=[sdata["Country"],sdata["State/Province"]]).count()))
# 当提前取出Series时，by要传入Series，三个会根据index合并，然后再分组。
# 相当于先取三列出来，再分组

print(list(sdata.groupby(by=["Country","State/Province"]))[0])
# 格式：((key1,key2),dataFrmae)

sdata["Brand"].groupby(by=[sdata["Country"],sdata["State/Province"]]).count()
# 两列index的Serices类型！！！！
```

```python
# 复合索引

df.set_index(["a","b"])
# 两列作为index
```

```python
# 取复合索引 Series的值
mutix_index_df = df.set_index(["a","b"])
print(mutix_index_df["c"]) # 先取一列为 Series
print(mutix_index_df["c"]["A"]) # 再根据第一个索引取出一个但索引Series
print(mutix_index_df["c"]["A"]["B"]) # 再根据第二个索引取

print()
print(mutix_index_df["c"]["A","B"]) # Series中取复合索引值的话，直接用数组即可。df不行，会被认为取多列

# 取复合索引 df的值
print()
print(mutix_index_df.loc["A"].loc["B"])  # 或者通过loc来取。要些两次loc。把一行变为一个Serries
print(mutix_index_df.loc["A"].loc["B"]["c"])  # 再根据索引取Series的指定位置的值
print(mutix_index_df.loc["A"].loc["B"][1])  # 再根据索引取Series的指定位置的值

```
