```python
# 一共有116年的数据
print(len(pop["year"].unique()))
# 两个州个50年的人口没有统计
```

```python
# 发现就是两个州 1900-1949年例没有人口数据
pop["year"][pop["pop"].isna()].unique()
```

```python
p1 = pop.dropna(axis=0).groupby("year").aggregate({
    "state": len,
    "pop": np.sum
})
```

```python
p_avg = p1["pop"]/p1["state"]
p_avg
```

```python
plt.figure(num=1, figsize=(20, 8), dpi=80)
plt.grid(alpha=0.5)
plt.rcParams['font.sans-serif'] = 'Microsoft YaHei'
plt.rcParams['axes.unicode_minus'] = False
plt.xlabel("年份")
plt.ylabel("平均人口")
plt.plot(p_avg.index, p_avg)
plt.xticks(range(np.min(p_avg.index), np.max(
    p_avg.index), 7), p_avg.index[::7])
plt.show()
```

```python
park_types = all_data.loc[:, ["Gnis Id", "Unit Type"]
                          ].groupby("Unit Type").count()
```

```python
park_types.head(3)
```

```python
plt.figure(num=1, figsize=(20, 8))
plt.rcParams['font.sans-serif'] = 'Microsoft YaHei'
plt.rcParams['axes.unicode_minus'] = False
x = range(len(park_types))
bar1 = plt.bar(x, park_types["Gnis Id"], width=0.8)
plt.grid(axis="y", alpha=0.4)
names = park_types.index.str.replace(" ", "\n")

for b in bar1:
    height = b.get_height()
    plt.text(b.get_x()+b.get_width()/2, height+1,
             str(height), ha='center', va='bottom')

plt.ylim(0, 6000)
plt.show()
```

```python
from mpl_toolkits.basemap import Basemap
```

```python
def point_park(a,m):
    xpt, ypt = m(a[0], a[1])
    m.plot(xpt, ypt, 'b^', markersize=3)

plt.figure(figsize=(20,8))
m = Basemap(projection='mill',
            llcrnrlat = 15,
            llcrnrlon = -180,
            urcrnrlat = 75,
            urcrnrlon = -20,
            resolution='l')
# llcrnrlat - 左下角的纬度
# llcrnrlon - 左下角的经度
# urcrnrlat - 右上角的纬度
# urcrnrlon - 右上角的经度
# 此外，坐标需要转换，其中西经和南纬坐标是负值，北纬和东经坐标是正值。
m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawcounties(color='darkred')
# m.etopo()
locations.loc[:,["lon","lat"]].apply(point_park,axis=1,args=(m,))
plt.show()
```
